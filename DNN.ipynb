```python
import tensorflow as tf
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split

# Load the dataset
df = pd.read_csv('file.csv')

# Drop the 'year' and 'Month' columns
df = df.drop(['year', 'Month'], axis=1)

# Split the dataset into train and test sets
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# Normalize the train and test sets
train_df_norm = (train_df - train_df.mean()) / train_df.std()
test_df_norm = (test_df - train_df.mean()) / train_df.std()

# Define the DNN model
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(3,)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1)
])

# Compile the model
model.compile(loss='mean_absolute_error', optimizer='adam')

# Train the model
model.fit(train_df_norm.drop('Rainfall', axis=1), train_df_norm['Rainfall'], epochs=50, batch_size=32, validation_split=0.2)

# Evaluate the model on the train and test sets
train_pred = model.predict(train_df_norm.drop('Rainfall', axis=1))
train_nmae = (train_df_norm['Rainfall'] - train_pred.reshape(-1)) / train_df_norm['Rainfall'].std()
train_rnmse = (sum((train_df_norm['Rainfall'] - train_pred.reshape(-1)) ** 2) / len(train_pred)) ** 0.5 / train_df_norm['Rainfall'].std()

test_pred = model.predict(test_df_norm.drop('Rainfall', axis=1))
test_nmae = (test_df_norm['Rainfall'] - test_pred.reshape(-1)) / test_df_norm['Rainfall'].std()
test_rnmse = (sum((test_df_norm['Rainfall'] - test_pred.reshape(-1)) ** 2) / len(test_pred)) ** 0.5 / test_df_norm['Rainfall'].std()
nmae_avg = np.mean([train_nmae, test_nmae])
nrmse_avg = np.mean([train_rnmse, test_rnmse])

# Print the NMAE and RNMSE values for the train and test sets
print('Train NMAE:', train_nmae.abs().mean())
print('Train RNMSE:', train_rnmse)
print('Test NMAE:', test_nmae.abs().mean())
print('Test RNMSE:', test_rnmse)
```
